```python
import json
import os
import pandas as pd
import numpy as np
from datetime import datetime

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

### Weather data process

```python
json_files = [i for i in os.listdir('data/') if 'json' in i]
```

```python
json_files_list = {}

for i in json_files:
    path = 'data/' + i
    city = i.split('.')[0]
    f = open(path)
    data = json.load(f)
    json_files_list[city] = data
    
data_list = []

for city in json_files_list.keys():
    data = pd.DataFrame(json_files_list[city])
    data['city'] = city
    data_list.append(data)
```

```python
weather = pd.concat(data_list)
weather.drop(weather.isnull().sum()[weather.isnull().sum() > 20000].index, 1, inplace = True)

weather['utc_time'] = weather['time'].apply(lambda x: datetime.utcfromtimestamp(x))
weather.drop(['icon', 'summary', 'time'], 1, inplace = True)

weather_city_mapper = {'san_antonio' : 'san antonio', 'san_diego' : 'san diego', 'san_jose' : 'san jose'}
weather['city'] = weather['city'].replace(weather_city_mapper)
```

```python
weather.describe(exclude = ['O', 'datetime'], percentiles = [.01, .1, .25, .5, .75, .9, .99]).T
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
sns.heatmap(weather.corr(), 
            ax = ax, cmap = 'vlag', center = 0, annot = True, fmt = '.2g')
plt.title('Correlations are mostly weak\n')
plt.show()
```

```python
weather.drop(['windGust', 'temperature'], 1, inplace = True)
```

### Electricity data

```python
[i for i in os.listdir('data/') if 'csv' in i]
```

```python
balance = pd.read_csv('data/cleaned_balance_data.csv')
balance.drop(['company'], 1, inplace = True)
balance.dropna(inplace = True)

balance['local_time'] = pd.to_datetime(balance['local_time'])
balance['utc_time'] = pd.to_datetime(balance['utc_time'])


subreg = pd.read_csv('data/cleaned_subregion_data.csv')
subreg.drop(['company', 'region'], 1, inplace = True)
subreg.dropna(inplace = True)

subreg['local_time'] = pd.to_datetime(subreg['local_time'])
subreg['utc_time'] = pd.to_datetime(subreg['utc_time'])


# texas = pd.read_csv('data/cleaned_texas_data.csv')
# texas = texas.melt(id_vars = 'date', var_name = 'city', value_name = 'demand')

# texas['local_time'] = pd.to_datetime(texas['date'])
# texas.drop(['date'], 1, inplace = True)
```

```python
data = pd.concat([balance, subreg])
```

```python
data.describe(exclude = ['O', 'datetime'], percentiles = [.01, .1, .25, .5, .75, .9, .99]).T
```

```python
data.loc[data['demand'] > 30000, 'demand'] = np.nan
data['demand'] = data['demand'].interpolate('linear')
data = data[data['demand'] > 0]
```

```python
data.describe(exclude = ['O', 'datetime'], percentiles = [.01, .1, .25, .5, .75, .9, .99]).T
```

```python
data = data.merge(weather, on = ['city', 'utc_time'])
```

## Temperature - electricity demand relationship

```python
# noticed outliers in Seattle

drop_indices = data[(data['city'] == 'seattle') & (data['demand'] > 6000) | (data['demand'] < 400)].index
data.drop(drop_indices, inplace = True)
```

Too many datapoints (hourly data), so let's take daily aggregates

```python
city_date_agg = data.groupby([data['city'], data['utc_time'].dt.date]).agg({'apparentTemperature' : 'mean', 'demand' : 'sum'}).reset_index()
city_date_agg.head(2)
```

```python
fig = px.scatter(data_frame = city_date_agg, y = 'demand', x = 'apparentTemperature', hover_data = ['utc_time'], 
                 facet_row_spacing = 0.075, facet_col_spacing = 0.05,
                  title = 'Temperature vs electricity consumed', facet_col = 'city', facet_col_wrap = 3,
                  labels = {'apparentTemperature' : 'Temperature in F', 'demand' : 'MW'},
                  width = 950, height = 850)

fig.update_xaxes(visible = True, showticklabels = True, title = None, matches = None)
fig.update_yaxes(visible = True, showticklabels = True, title = None, matches = None)
fig.for_each_annotation(lambda a: a.update(text = a.text.split('=')[-1]))
fig.update_traces(marker = {'size' : 4}, line = {'color' : 'black', 'width' : 1})
fig.show()
```

Temperature has a quadratic effect on electricity demand - as expected. Above and below a certain temperature level, the demand for electricity starts to increase (for cooling, or heating)
