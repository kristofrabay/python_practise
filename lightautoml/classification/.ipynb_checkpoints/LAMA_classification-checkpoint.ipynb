## Running LightAutoML for classification problem
Will use the same dataset as was used for AutoWoE and see by how much the whitebox (WoE, logreg) AUC (0.8459) can be improved

```python
import calendar
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split

from lightautoml.automl.presets.tabular_presets import TabularAutoML, TabularUtilizedAutoML
from lightautoml.tasks import Task
from lightautoml.report.report_deco import ReportDeco

import warnings
warnings.filterwarnings('ignore')

def plot_feat_imp(feat_scores_object):

    plt.figure(figsize = (9, 4))
    sns.barplot(data = feat_scores_object, x = 'Importance', y = 'Feature', 
                color = '#89bedc', edgecolor = 'black', linewidth = 1/2)
    plt.title('LightAutoML feature importances')
    plt.xlabel(None)
    plt.ylabel(None)
    plt.show()
    
RANDOM_STATE = 56
```

### Data: binary output 1 if daily bike shares reached 6000, 0 otherwise

```python
data = pd.read_csv('../regression/data/day.csv', index_col = 0)
data.reset_index(drop = True, inplace = True)
data.drop(['dteday', 'yr', 'casual', 'registered', 'temp'], 1, inplace = True)

season_mapper = {1 : 'Spring', 2 : 'Summer', 3 : 'Fall', 4 : 'Winter'}
data['season'].replace(season_mapper, inplace = True)

month_mapper = dict(zip([i for i in range(0, 13)], [i for i in calendar.month_abbr]))
data['mnth'].replace(month_mapper, inplace = True)

day_mapper = dict(zip([1, 2, 3, 4, 5, 6, 0], [i for i in calendar.day_abbr]))
data['weekday'].replace(day_mapper, inplace = True)

weather_mapper = {1 : 'Clear', 2 : 'Cloudy', 3 : 'Light Precip', 4 : 'Heavy Precip'}
data['weathersit'].replace(weather_mapper, inplace = True)

data['reached_6000'] = data['cnt'].apply(lambda x: 1 if x > 6000 else 0)
# data['reached_6000'] = data['reached_6000'].astype('category')
data.drop('cnt', 1, inplace = True)

data['reached_6000'].value_counts() / data.shape[0]
```

```python
data.head(3)
```

### Create LightAutoML classifier

```python
TARGET = 'reached_6000'
```

Set categoricals

```python
categoricals = data.select_dtypes('O').columns.tolist()
print(categoricals)
```

Split to train and test

```python
train, test = train_test_split(data, train_size = .8, random_state = RANDOM_STATE, stratify = data[TARGET])
```

#### Default model

```python
%%time

RD = ReportDeco(output_path = 'TabularAutoML_Report')

N_THREADS = 4
N_FOLDS = 5
TIMEOUT = 600

task = Task('binary', loss = 'logloss', metric = 'auc')

roles = {'target': TARGET,
         'category' : categoricals}

automl = TabularAutoML(task = task, 
                       cpu_limit = N_THREADS,
                       reader_params = {'n_jobs': N_THREADS, 'cv': N_FOLDS, 'random_state': RANDOM_STATE},
                       general_params = {'use_algos': [['linear_l2', 'lgb', 'lgb_tuned']]})

automl_rd = RD(automl)

oof_pred = automl_rd.fit_predict(train, roles = roles)
```

```python
predictions = automl_rd.predict(test).data[:,0]
actuals = test['reached_6000']
```

```python
print('AUC on test:', roc_auc_score(actuals, predictions).round(5))
```

AutoWhiteBoxML's woe based logreg beat ordinary logistic regression, and now AutoML beats WhiteBox's solution, as AUC increased from 0.8459 to 0.8728

Now open the model and check some stats:

```python
plot_feat_imp(automl_rd.model.get_feature_scores())
```

```python
print(automl_rd.model.create_model_str_desc())
```

```python
automl_rd.model.plot_pdp(test, feature_name = 'atemp')
```

#### Utilized model with longer train time and more models

```python
automl_long = TabularUtilizedAutoML(task = task, 
                               timeout = 60 * 10,
                               cpu_limit = N_THREADS,
                               reader_params = {'n_jobs': N_THREADS, 'cv': N_FOLDS, 'random_state': RANDOM_STATE},
                               general_params = {'use_algos': [['linear_l2', 'lgb', 'lgb_tuned', 'cb'], ['linear_l2', 'lgb']]})
```

```python
%%time
oof_pred = automl_long.fit_predict(train, roles = roles)
```

```python
predictions_UT = automl_long.predict(test).data[:,0]
actuals_UT = test['reached_6000']

print('AUC on test with utilized model:', roc_auc_score(actuals_UT, predictions_UT).round(5))
```

AUC increased from 87.3 to 89 by adding CatBoost to first level, setting a second level, and training for 10 minutes
