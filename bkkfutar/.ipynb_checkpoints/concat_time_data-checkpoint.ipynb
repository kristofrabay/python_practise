```python
import pandas as pd
import plotly.express as px
import os

import warnings
warnings.filterwarnings('ignore')
```

```python
collection = []

for table in os.listdir('vehicle_data/'):
    
    date_id = table[:-4]
    table_read_in = pd.read_csv('vehicle_data/' + table)
    table_read_in['datetime'] = date_id
    collection.append(table_read_in)    
```

```python
data = pd.concat((df for df in collection), axis = 0, join = 'outer')
```

```python
data['datetime'] = pd.to_datetime(data['datetime'], format = '%Y-%m-%d %H-%M-%S')
data['datetime'] = data['datetime'].astype(str)
```

```python
def plot_vehicles_map():
    
    data['size'] = 1

    fig = px.scatter_mapbox(data, lat = 'lat', lon = 'lon', hover_name = 'shortName',
                            hover_data = ['shortName', 'licensePlate', 'model', 'vehicleRouteType', 'From', 'To'],
                            title = 'Forgalomban lévő járművek alakulása', size = 'size', size_max = 4.5, 
                            labels = {'licensePlate' : 'Rendszám', 'shortName' : 'Jármű száma',
                                      'model' : 'Típus', 'vehicleRouteType' : 'Járműtípus', 'datetime' : 'Idő'},
                            color = 'vehicleRouteType', animation_frame = 'datetime',
                            color_discrete_map = {'BUS' : 'blue', 'TRAM' : 'yellow', 'TROLLEYBUS' : 'red', 'SUBURBAN_RAILWAY' : 'green'},
                            zoom = 10, mapbox_style = 'carto-positron', height = 650, width = 1050)
    
    fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 100
    fig.show()
    
    data.drop('size', 1, inplace = True)
```

```python
plot_vehicles_map()
```
