### 3. word2vec + logit

```python
%%time 

from gensim.models import KeyedVectors
from gensim.models import Word2Vec

EMBEDDING_FILE = '../../../../../Downloads/GoogleNews-vectors-negative300.bin.gz'
google_word2vec = KeyedVectors.load_word2vec_format(EMBEDDING_FILE, binary = True)

def w2v_vectorize_line(line):
    
    # assign vectors to words    
    vectors = []    
    
    for word in line.split():                   
        if word not in google_word2vec.index_to_key:
            vectors.append(np.zeros(300))            
        else:
            vectors.append(google_word2vec[word])
            
    # calculate avg    
    n = len(vectors)
    N = np.zeros(300)    
    
    for vector in vectors:        
        N += vector
        
    avg_vector = N / n  
    
    # return as list
    return avg_vector
```

Apply word2vec vectorizer and store in array

```python
train_w2v_list = train['Comment_cleaned_non_BERT'].tolist()
test_w2v_list = test['Comment_cleaned_non_BERT'].tolist()
```

```python
%%time

train_w2v = [w2v_vectorize_line(line) for line in train_w2v_list]
test_w2v = [w2v_vectorize_line(line) for line in test_w2v_list]
```

```python
train_w2v = np.array(train_w2v)
test_w2v = np.array(test_w2v)

# if text is missing (due to cleaning) replace null vector with 0 vector
train_w2v[np.isnan(train_w2v)] = 0
test_w2v[np.isnan(test_w2v)] = 0

print(train_w2v.shape)
print(test_w2v.shape)
```

### 4. word2vec as Embedding layer in neural net

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

Tokenize words

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_w2v_list)
```

```python
train_w2v_tokenized = tokenizer.texts_to_sequences(train_w2v_list)
test_w2v_tokenized = tokenizer.texts_to_sequences(test_w2v_list)
```

```python
train_w2v_list[12]
```

```python
train_w2v_tokenized[12]
```

Pad sequences

```python
print('mean length: ', np.mean([len(text) for text in train_w2v_tokenized]))
print('max length: ', np.max([len(text) for text in train_w2v_tokenized]))

max_len = 35
print('\nWill set the max length to', max_len)
```

```python
train_w2v_tokenized_padded = pad_sequences(train_w2v_tokenized, maxlen = max_len, padding = 'pre')
test_w2v_tokenized_padded = pad_sequences(test_w2v_tokenized, maxlen = max_len, padding = 'pre')

print('Shape of padded sentences in train:', train_w2v_tokenized_padded.shape)
print('Shape of padded sentences in test:', test_w2v_tokenized_padded.shape)
```

Create Embedding dict

```python
vocab_size = len(tokenizer.word_index) + 1

embedding_matrix = np.zeros((vocab_size, 300))

for word, i in tokenizer.word_index.items():
    
    if word not in google_word2vec.index_to_key:
        embedding_vector = np.zeros(300)           
    else:
        embedding_vector = google_word2vec[word]
        
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
```

```python
embedding_matrix.shape
```

Train simple neural net

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Embedding, Flatten, Convolution1D, MaxPooling1D
from tensorflow.keras.optimizers import Adam

from keras.utils.np_utils import to_categorical
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
from livelossplot import PlotLossesKeras
```

```python
print('One hot encoding target:', train['Topic'].factorize()[1])

y_train = to_categorical(np.asarray(train['Topic'].factorize()[0]))
y_test = to_categorical(np.asarray(test['Topic'].factorize()[0]))
```

```python
model = Sequential(name = 'w2v_in_embedding_layer')

model.add(Embedding(vocab_size, output_dim = embedding_matrix.shape[1], 
                    weights = [embedding_matrix], input_length = max_len, trainable = False))
model.add(Dropout(rate = 1/4))

model.add(Convolution1D(filters = 100, kernel_size = 5, padding = 'valid', activation = 'relu', strides = 1))
model.add(Convolution1D(filters = 100, kernel_size = 5, padding = 'valid', activation = 'relu', strides = 1))

model.add(MaxPooling1D())

model.add(Flatten())

model.add(Dense(128, activation = 'relu'))
model.add(Dropout(rate = 1/4))

model.add(Dense(y_train.shape[1], activation='softmax'))

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
np.random.seed(20202020)

callbacks = [PlotLossesKeras(),
             EarlyStopping(monitor = 'val_loss', patience = 5),
             ReduceLROnPlateau(monitor = 'val_loss', patience = 3, factor = 2/3)]

model.fit(train_w2v_tokenized_padded, y_train, batch_size = 32, epochs = 30, 
          validation_split = 0.2, verbose = 1, callbacks = callbacks, workers = 4)
```

Evaluate on test

```python
w2v_net_accuracy, w2v_net_report = evaluate_text_class_predictions(test['Topic'], model, test_w2v_tokenized_padded, 'word2vec in simple neural net', 
                                                                       neural_net = True)
```

### 8. BERT on original text + logit

```python
import torch
from transformers import DistilBertTokenizer, DistilBertModel

tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
model = DistilBertModel.from_pretrained("distilbert-base-uncased")
```

Tokenize and pad train and test sequences

```python
def bert_tokenizer(text, MAX_LEN = 50):
    
    encoded_text = tokenizer.encode_plus(text,  max_length = MAX_LEN, truncation=True,  padding='max_length',  
                                         return_attention_mask=True, return_tensors='pt')
    
    return encoded_text['input_ids'][0].numpy(), encoded_text['attention_mask'][0].numpy()
```

```python
train_tokenized, train_masks = zip(*train['Comment_cleaned_BERT'].apply(lambda x: bert_tokenizer(x)))
test_tokenized, test_masks = zip(*test['Comment_cleaned_BERT'].apply(lambda x: bert_tokenizer(x)))
```

```python
train_tokenized = torch.tensor(train_tokenized)  
train_masks = torch.tensor(train_masks)

test_tokenized = torch.tensor(test_tokenized)  
test_masks = torch.tensor(test_masks)

print(train_tokenized.shape)
print(train_masks.shape)

print(test_tokenized.shape)
print(test_masks.shape)
```

Apply BERT on tokens

```python
%%time

with torch.no_grad(): # no need to keep track of gradients
    train_last_hidden_states = model(train_tokenized, attention_mask = train_masks)
```

```python
%%time

with torch.no_grad(): # no need to keep track of gradients
    test_last_hidden_states = model(test_tokenized, attention_mask = test_masks)
```

### 9. BERT fine tuned via neural net

```python
from tensorflow import config as tf_config
tf_config.run_functions_eagerly(True)
```

```python
from transformers import TFDistilBertForSequenceClassification

from tensorflow.keras.callbacks import ModelCheckpoint
checkpoint_path = "bert_checkpoint/distilbert_ckpts.ckpt"
model_callback = ModelCheckpoint(filepath = checkpoint_path, save_weights_only=True, verbose=1)

from tensorflow.data import Dataset
from tensorflow.keras.optimizers import Adam
from keras.losses import SparseCategoricalCrossentropy

model = TFDistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels = 3)
```

```python
BATCH_SIZE = 32
```

```python
mapper = {'Biology' : 0, 'Physics' : 1, 'Chemistry' : 2}

y_train = train['Topic'].replace(mapper).tolist()
y_test = test['Topic'].replace(mapper).tolist()
```

```python
def construct_encodings(x, tokenizer = tokenizer,  MAX_LEN = 50, trucation=True, padding=True):
    return tokenizer(x, max_length = MAX_LEN, truncation=trucation, padding=padding)
    
train_encodings = construct_encodings(train['Comment_cleaned_BERT'].tolist())
test_encodings = construct_encodings(test['Comment_cleaned_BERT'].tolist())
```

```python
def construct_tfdataset(encodings, y=None):
    if y:
        return Dataset.from_tensor_slices((dict(encodings),y))
    else:
        # this case is used when making predictions on unseen samples after training
        return Dataset.from_tensor_slices(dict(encodings))
    
train_bert = construct_tfdataset(train_encodings, y_train).batch(BATCH_SIZE)
test_bert = construct_tfdataset(test_encodings, y_test).batch(BATCH_SIZE)
```

```python
model.layers[0].trainable = False
model.layers[1].trainable = False
```

```python
print(model.summary())
```

```python
optimizer = Adam(learning_rate=3e-5)
loss = SparseCategoricalCrossentropy(from_logits = True)

model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
```

```python
np.random.seed(20202020)

history = model.fit(train_bert, batch_size=BATCH_SIZE, epochs=2)
```

Evaluate on test

```python
model.evaluate(test_bert, return_dict=True, batch_size = BATCH_SIZE)
```

```python
from keras.activations import softmax
from tensorflow import convert_to_tensor

from sklearn.metrics import accuracy_score, classification_report
```

```python
test_logits = model.predict(test).logits
```

```python
classes = softmax(convert_to_tensor(test_logits)).numpy().argmax(axis=-1)
```

```python
predictions_classes = list(map({0 : 'Biology', 1 : 'Physics', 2 : 'Chemistry'}.get, classes))

BERT_tuned_accuracy = accuracy_score(test['Topic'], predictions_classes)
BERT_tuned_report = classification_report(test['Topic'], predictions_classes)
```
