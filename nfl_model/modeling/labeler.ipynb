```python
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../data/scraped_for_modeling.csv')
data.head(3)
```

```python
winner = {2004 : 'New England Patriots', 2005 : 'Pittsburgh Steelers', 2006 : 'Indianapolis Colts', 
          2007 : 'New York Giants', 2008 : 'Pittsburgh Steelers', 2009 : 'New Orleans Saints', 
          2010 : 'Green Bay Packers', 2011 : 'New York Giants', 2012 : 'Baltimore Ravens', 
          2013 : 'Seattle Seahawks', 2014 : 'New England Patriots', 2015 : 'Denver Broncos', 
          2016 : 'New England Patriots', 2017 : 'Philadelphia Eagles', 2018 : 'New England Patriots', 
          2019 : 'Kansas City Chiefs', 2020 : 'Tampa Bay Buccaneers'}

loser = {2004 : 'Philadelphia Eagles', 2005 : 'Seattle Seahawks', 2006 : 'Chicago Bears', 
         2007 : 'New England Patriots', 2008 : 'Arizona Cardinals', 2009 : 'Indianapolis Colts', 
         2010 : 'Pittsburgh Steelers', 2011 : 'New England Patriots', 2012 : 'San Francisco 49ers', 
         2013 : 'Denver Broncos', 2014 : 'Seattle Seahawks', 2015 : 'Carolina Panthers', 
         2016 : 'Atlanta Falcons', 2017 : 'New England Patriots', 2018 : 'Los Angeles Rams', 
         2019 : 'San Francisco 49ers', 2020 : 'Kansas City Chiefs'}
```

```python
data['winner'] = data['season'].map(winner)
data['loser'] = data['season'].map(loser)
```

```python
data['winner'] = data.apply(lambda x: 1 if x['winner'] == x['Team'] else 0, 1)
data['played'] = data.apply(lambda x: 1 if (x['loser'] == x['Team']) or (x['winner'] == 1) else 0, 1)
data.drop(['loser'], 1, inplace = True)
```

```python
data.to_csv('../data/scraped_for_modeling_labeled.csv', index = False)
```
