```python
import pandas as pd
import numpy as np

from datetime import datetime
from ast import literal_eval
from collections import Counter, OrderedDict

import matplotlib.pyplot as plt
import seaborn as sns
    
import warnings
warnings.filterwarnings('ignore')
```

```python
data_link = 'data/archive.zip'
data = pd.read_csv(data_link, compression = 'zip')
print(data.shape)
```

### Dropping unnecessary / unusable columns

```python
unnamed_cols = data.filter(like = 'Unnamed').columns
data.drop(unnamed_cols, 1, inplace = True)
```

```python
nan_check = data.isnull().sum() / data.shape[0]
drop_nan = nan_check[nan_check > 1/3].index.tolist()

print('Dropping: ', drop_nan)
data.drop(drop_nan, 1, inplace = True)
```

```python
unnecessary_cols = ['id', 'prices.sourceURLs', 'asins', 'imageURLs', 'keys', 'name', 'primaryCategories', 
                    'manufacturerNumber', 'upc', 'sourceURLs', 'prices.dateSeen', 'prices.availability']
data.drop(unnecessary_cols, 1, inplace = True)
```

### Converting features to correct dtypes

```python
data['dateAdded'] = data['dateAdded'].apply(lambda x: datetime.strptime(x, '%Y-%m-%dT%H:%M:%S%fZ').date())
data['dateUpdated'] = data['dateUpdated'].apply(lambda x: datetime.strptime(x, '%Y-%m-%dT%H:%M:%S%fZ').date())
```

Fix weights

```python
pounds = ['pounds', 'lb', 'lbs', 'lb.', 'Pounds']
ounces = ['ounces', 'oz', 'OUNCE']
kg = ['Kg']
g = ['g']

data['weight_measure'] = np.nan
data['weight_measure'] = data['weight'].apply(lambda x: 'pounds' if any(i in x for i in pounds) else 
                                                        ('ounces' if any(i in x for i in ounces) else
                                                        ('kg' if any(i in x for i in kg) else
                                                        ('g' if any(i in x for i in g) else np.nan))))
print(data.shape)
data = data[data['weight_measure'].notnull()]
print(data.shape)
```

```python
data['weight_length'] = data['weight'].apply(lambda x: len(x))
data['weight_length'].describe(percentiles = [.1, .75, .9, .99])
```

```python
print(data.shape)
data = data[data['weight_length'] <= 116] # long name is due to some scraping error, getting some html tags
print(data.shape)
```

```python
data['weight'] = data['weight'].str.replace('[a-zA-Z]', '').str.split().str[0].astype(float)
```

```python
conversion_dict = {'pounds' : 16, 'g' : 0.0352739619, 'kg' : 35.2739619, 'ounces' : 1}
data['weight_ounces'] = data['weight_measure'].map(conversion_dict) * data['weight']

data.drop(['weight_measure', 'weight', 'weight_length'], 1, inplace = True)
```

```python
data['prices.condition'] = data['prices.condition'].str.lower()

condition_dict = {'seller refurbished' : 'refurbished', 'manufacturer refurbished' : 'refurbished', 'new' : 'new',
                  'pre-owned' : 'used', 'new other (see details)' : 'new', 'brand new' : 'new', 'used' : 'used'}

data['prices.condition'] = data['prices.condition'].map(condition_dict)
data['prices.condition'].fillna('missing', inplace = True)
```

```python
data['prices.merchant'] = data['prices.merchant'].str.lower()
data['prices.merchant'].fillna('missing', inplace = True)

data['prices.merchant'] = data['prices.merchant'].apply(lambda x: 'bestbuy' if 'best' in x else x)
data['prices.merchant'] = data['prices.merchant'].apply(lambda x: 'amazon' if 'amazon' in x else x)
data['prices.merchant'] = data['prices.merchant'].apply(lambda x: 'walmart' if 'wal' in x else x)
data['prices.merchant'] = data['prices.merchant'].apply(lambda x: 'bhphotovideo' if 'bhphotovideo' in x else x)
data['prices.merchant'] = data['prices.merchant'].apply(lambda x: x if any(i in x for i in ['walmart', 'amazon', 'bestbuy', 'missing', 'bhphotovideo']) else 'other')
```

```python
conversion_dict = {'CAD' : 0.8, 'EUR' : 1.19, 'SGD' : 0.75, 'GBP' : 1.38, 'USD' : 1}

data['prices.amountMax'] = data['prices.currency'].map(conversion_dict) * data['prices.amountMax']
data['prices.amountMin'] = data['prices.currency'].map(conversion_dict) * data['prices.amountMin']

data.drop('prices.currency', 1, inplace = True)
```

```python
data['brand'] = data['brand'].str.lower()
top_brands = data['brand'].value_counts().head(5).index

data['brand'] = data['brand'].apply(lambda x: 'other' if x not in top_brands else x)
```

```python
def consolidate_amenities_lists(amenities_list):
    
    consolidated_list = ['computers' if 'computer' in str(i).lower() else i for i in amenities_list]
    consolidated_list = ['audio' if ('audio' in str(i).lower() or 
                                     'stereo' in str(i).lower() or
                                     'speaker' in str(i).lower() or
                                     'headphone' in str(i).lower() or
                                     'speaker' in str(i).lower()) else i for i in consolidated_list]
    consolidated_list = ['tv' if 'tv' in str(i).lower() else i for i in consolidated_list]
    consolidated_list = ['tv' if 'television' in str(i).lower() else i for i in consolidated_list]
    consolidated_list = ['car' if ('car' in str(i).lower() or
                                   'auto' in str(i).lower()) else i for i in consolidated_list]
    consolidated_list = ['drives' if ('drive' in str(i).lower() or
                                      'ssd' in str(i).lower()) else i for i in consolidated_list]
    consolidated_list = ['electronics' if 'electronic' in str(i).lower() else i for i in consolidated_list]
    consolidated_list = ['camera' if ('camera' in str(i).lower() or
                                      'photo' in str(i).lower()) else i for i in consolidated_list]
    consolidated_list = ['mobile' if ('cell' in str(i).lower() or
                                      'phone' in str(i).lower() or
                                      'mobil' in str(i).lower()) else i for i in consolidated_list]
    consolidated_list = ['laptop' if 'laptop' in str(i).lower() else i for i in consolidated_list]
    consolidated_list = ['storage' if 'storage' in str(i).lower() else i for i in consolidated_list]
    consolidated_list = ['accessories' if 'access' in str(i).lower() else i for i in consolidated_list]
        
    consolidated_list = list(OrderedDict.fromkeys(consolidated_list))
    
    return consolidated_list
```

```python
data['categories'] = data['categories'].apply(lambda x: consolidate_amenities_lists(x))
```

```python
len(set().union(*data['categories']))
```

```python
flattened_list = [item for sublist in data['categories'].tolist() for item in sublist]
cat_count = Counter(flattened_list).most_common(10)
top_cat = [i[0] for i in cat_count]

print('Top categories included: ', top_cat)
```

```python
def reduce_categories(cat_list):    
    reduced_categories = ['category_' + '_'.join(i.split()).lower() for i in cat_list if i in top_cat]    
    return sorted(reduced_categories)
```

```python
data['categories'] = data['categories'].apply(lambda x: reduce_categories(x))
```

```python
for col in list(set().union(*data['categories'])):       
    data[col] = data['categories'].apply(lambda x: 1 if col in x else 0)
    
data.drop(['categories'], 1, inplace = True)
```

## Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

from catboost import CatBoostRegressor

def mean_absolute_percentage_error(actual, pred): 
    return np.mean(np.abs((actual - pred) / actual)) * 100

def evaluate_model(model_name, model, X, y):
    
    predictions = model.predict(X)

    MAE = mean_absolute_error(y, predictions)
    MAPE = mean_absolute_percentage_error(y, predictions)
    RMSE = mean_squared_error(y, predictions, squared = False)
    R2 = r2_score(y, predictions)

    print('MAE for', model_name, ': %1.3f' % MAE)
    print('MAPE for', model_name, ': %1.3f' % MAPE)
    print('RMSE for', model_name, ': %1.3f' % RMSE)
    print('R2 for', model_name, ': %1.3f' % R2)
    
    
def plot_predictions(model, x, y, title):
    
    plt.figure(figsize = (7, 4))
    sns.scatterplot(x = y, y = model.predict(x), s = 20)
    sns.lineplot([y.min(), y.max()], [y.min(), y.max()], color = 'black', linewidth = 1)
    plt.title(title + '\n')
    plt.xlabel('Actual values')
    plt.ylabel('Predicted values')
    plt.show()
```

```python
y = data['prices.final']
X = data.drop('prices.final', 1).copy()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 20202020, train_size = 0.8)
```

```python
categorical_features = X.select_dtypes(include = 'O').columns.tolist()
print(categorical_features)
```

```python
default_model = CatBoostRegressor(random_state = 20202020, 
                                  learning_rate = 0.05,
                                  # rsm = 1, min_data_in_leaf = 100, l2_leaf_reg = 5,
                                  iterations = 1000,
                                  # early_stopping_rounds = 15,
                                  cat_features = categorical_features,
                                  one_hot_max_size = 20,
                                  verbose = 100)
default_model.fit(X_train, y_train, eval_set = (X_test, y_test), plot = True, verbose = False)
```

```python
evaluate_model('Default CatBoost on train', default_model, X_train, y_train)
```

```python
evaluate_model('Default CatBoost on test', default_model, X_test, y_test)
```

```python
plot_predictions(default_model, X_train, y_train, 'Default CatBoost train performance')
```

```python
plot_predictions(default_model, X_test, y_test, 'Default CatBoost test performance')
```
