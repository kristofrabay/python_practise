```python
import pandas as pd
import numpy as np
from ast import literal_eval

import matplotlib.pyplot as plt
import seaborn as sns
#import plotly.express as px

import statsmodels.api as sm
from statsmodels.tools.eval_measures import rmse
from sklearn.feature_extraction.text import TfidfVectorizer
# from statsmodels.formula.api import ols
# from stargazer.stargazer import Stargazer

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from catboost import CatBoostRegressor, Pool

import shap

import warnings
warnings.filterwarnings('ignore')

def plot_dist(data, i, limit = np.inf):
    
    df = data[data[i] < limit].copy()
    
    plt.figure(figsize = (9, 4))
    sns.distplot(df[i], kde = False, bins = 40, hist_kws = {'edgecolor' : 'black', 'linewidth' : 1})
    plt.title('Distribution of ' + i)
    plt.xlabel(None)
    plt.ylabel(None)
    plt.show()  
    
def scatter(data, x, y):
    plt.figure(figsize = (9, 5))
    sns.scatterplot(data = data, x = x, y = y, size = 3/4, edgecolor = 'k', linewidth = 1/3, alpha = .75, color = 'blue', legend = False)
    plt.title('Relationship between ' + x + ' and ' + y)
    plt.show()  
    
def fit_sm_OLS(data):
    
    y_name = data.filter(like = 'Vétel').columns[0]
    
    Y = data[y_name].copy()
    X = data.drop(y_name, 1).copy()
    X = sm.add_constant(X)
    
    model = sm.OLS(Y, X)
    results = model.fit()
    
    return results

def OLS_results(ols_model, name_of_model = ''):
    
    p_ = ols_model.pvalues
    p_insignificant = p_[p_ > 0.05]
    p_insignificant = p_insignificant.index.tolist()

    r2 = ols_model.rsquared
    r2_adj = ols_model.rsquared_adj

    feat = len(p_.index.tolist())
    insig_feat = len(p_insignificant)
    
    return pd.DataFrame({'Model' : name_of_model, 'r2' : r2, 'r2 adj.' : r2_adj, 'features num' : feat, 
                         'insig. feat num' : insig_feat, 'insig. feat' : [', '.join(p_insignificant)]})


def mean_absolute_percentage_error(actual, pred): 
    return np.mean(np.abs((actual - pred) / actual)) * 100

def evaluate_model(model_name, model, X, y):
    
    predictions = model.predict(X)
    feat = model.n_features_in_

    #MAE = mean_absolute_error(y, predictions)
    #MAPE = mean_absolute_percentage_error(y, predictions)
    #RMSE = mean_squared_error(y, predictions, squared = False)
    R2 = r2_score(y, predictions)

    #print('MAE for', model_name, ': %1.3f' % MAE)
    #print('MAPE for', model_name, ': %1.3f' % MAPE)
    #print('RMSE for', model_name, ': %1.3f' % RMSE)
    #print('R2 for', model_name, ': %1.3f' % R2)
    
    return pd.DataFrame({'Model' : model_name, 'r2' : R2, 'features num' : feat}, index = [0])
    

def plot_predictions(model, x, y, title):
    
    plt.figure(figsize = (9, 5))
    sns.scatterplot(x = y, y = model.predict(x), s = 20, size = 3/4, edgecolor = 'k', linewidth = 1/3, color = 'blue', legend = False)
    sns.lineplot([y.min(), y.max()], [y.min(), y.max()], color = 'black', linewidth = 1)
    plt.title(title + '\n')
    plt.xlabel('Actual values')
    plt.ylabel('Predicted values')
    plt.show()
```

#### Create regression models

```python
regression_data = data.copy()
```

```python
categoricals_with_modellcsoport = ['Állapot', 'Kivitel', 'Ajtók száma', 'Üzemanyag', 'Hajtás','Sebességváltó fajtája',
                                   'Okmányok jellege', 'Kategória', 'Szín', 'Klíma fajtája', 
                                   'Műszaki vizsga érvényes', 'Modellcsoport modellezéshez']

categoricals_with_marka = ['Állapot', 'Kivitel', 'Ajtók száma', 'Üzemanyag', 'Hajtás','Sebességváltó fajtája','Márka',
                           'Okmányok jellege', 'Kategória', 'Szín', 'Klíma fajtája', 'Műszaki vizsga érvényes']

numerics_ln = ['Évjárat', 'Saját tömeg', 'Csomagtartó', 'Szélesség', 'Magasság', 'Hosszúság', 
               'Üzemanyagtank', 'Hasznos teher', 'Vegyes fogyasztás', 'Végsebesség', 'Szín - metál', 
               'Kilométeróra állása', 'Log Teljesítmény LE'] # reciprok

numerics = ['Évjárat', 'Saját tömeg', 'Csomagtartó', 'Szélesség', 'Magasság', 'Hosszúság', 
            'Üzemanyagtank', 'Hasznos teher', 'Vegyes fogyasztás', 'Végsebesség', 'Szín - metál', 
            'Kilométeróra állása', 'Teljesítmény LE'] # reciprok
```

Adding text features as one hot encoded variables

```python
felszereltseg = ['Beltér', 'Műszaki', 'Kültér', 'Multimédia / Navigáció']
```

```python
%%time 

dummified_features = []

for i in felszereltseg:    
    
    regression_data[i] = regression_data[i].apply(lambda x: list(set(literal_eval(x))))
    i_dummified = regression_data[i].str.join('|').str.get_dummies()
    i_dummified.columns = [i + ' = ' + col for col in i_dummified.columns]
    dummified_features.append(i_dummified)
    
felszereltseg_df = pd.concat(dummified_features, 1)
```

```python
# felszereltseg_corr = felszereltseg_df.corr()
# felszereltseg_corr = felszereltseg_corr.mask(np.tril(np.ones(felszereltseg_corr.shape)).astype(np.bool))
# felszereltseg_corr[abs(felszereltseg_corr) >= 0.7].stack().reset_index()
```

```python
regression_data = regression_data.join(felszereltseg_df)
```

```python
felszereltseg_features = felszereltseg_df.columns.tolist()
```

Adding text features as vectorized variables

```python
for i in felszereltseg:
    regression_data[i] = regression_data[i].apply(lambda x: str(x).replace('[', '').replace(']', '').replace("'", '').strip())
```

```python
%%time

vectorized_text_dfs = []

for i in felszereltseg:
    
    vectorizer = TfidfVectorizer(max_features = 5)
    X = vectorizer.fit_transform(regression_data[i]).toarray()
    X = pd.DataFrame(X)
    X.columns = [i + ' feature ' + str(col) for col in X.columns]
    vectorized_text_dfs.append(X)
    
felszereltseg_df_vect = pd.concat(vectorized_text_dfs, 1)
felszereltseg_df_vect.drop(['Multimédia / Navigáció feature 0', 'Műszaki feature 0'], 1, inplace = True)
```

```python
regression_data = regression_data.join(felszereltseg_df_vect)
```

```python
felszereltseg_features_vect = felszereltseg_df_vect.columns.tolist()
```

Putting together model dataframes

```python
%%time

# without text

simple_model_with_brand = regression_data[categoricals_with_marka + numerics + ['Vételár']]
log_model_with_brand = regression_data[categoricals_with_marka + numerics_ln + ['Log Vételár']]
simple_model_with_modellcsoport = regression_data[categoricals_with_modellcsoport + numerics + ['Vételár']]
log_model_with_modellcsoport = regression_data[categoricals_with_modellcsoport + numerics_ln + ['Log Vételár']]

simple_model_with_brand = pd.get_dummies(simple_model_with_brand, columns = categoricals_with_marka, prefix_sep = ' = ')
simple_model_with_modellcsoport = pd.get_dummies(simple_model_with_modellcsoport, columns = categoricals_with_modellcsoport, prefix_sep = ' = ')
log_model_with_brand = pd.get_dummies(log_model_with_brand, columns = categoricals_with_marka, prefix_sep = ' = ')
log_model_with_modellcsoport = pd.get_dummies(log_model_with_modellcsoport, columns = categoricals_with_modellcsoport, prefix_sep = ' = ')

# with text

simple_model_with_brand_text = regression_data[categoricals_with_marka + numerics + felszereltseg_features + ['Vételár']]
log_model_with_brand_text = regression_data[categoricals_with_marka + numerics_ln + felszereltseg_features + ['Log Vételár']]
simple_model_with_modellcsoport_text = regression_data[categoricals_with_modellcsoport + numerics + felszereltseg_features + ['Vételár']]
log_model_with_modellcsoport_text = regression_data[categoricals_with_modellcsoport + numerics_ln + felszereltseg_features + ['Log Vételár']]

simple_model_with_brand_text = pd.get_dummies(simple_model_with_brand_text, columns = categoricals_with_marka, prefix_sep = ' = ')
simple_model_with_modellcsoport_text = pd.get_dummies(simple_model_with_modellcsoport_text, columns = categoricals_with_modellcsoport, prefix_sep = ' = ')
log_model_with_brand_text = pd.get_dummies(log_model_with_brand_text, columns = categoricals_with_marka, prefix_sep = ' = ')
log_model_with_modellcsoport_text = pd.get_dummies(log_model_with_modellcsoport_text, columns = categoricals_with_modellcsoport, prefix_sep = ' = ')

# with text as vectors

simple_model_with_brand_text_vect = regression_data[categoricals_with_marka + numerics + felszereltseg_features_vect + ['Vételár']]
log_model_with_brand_text_vect = regression_data[categoricals_with_marka + numerics_ln + felszereltseg_features_vect + ['Log Vételár']]
simple_model_with_modellcsoport_text_vect = regression_data[categoricals_with_modellcsoport + numerics + felszereltseg_features_vect + ['Vételár']]
log_model_with_modellcsoport_text_vect = regression_data[categoricals_with_modellcsoport + numerics_ln + felszereltseg_features_vect + ['Log Vételár']]

simple_model_with_brand_text_vect = pd.get_dummies(simple_model_with_brand_text_vect, columns = categoricals_with_marka, prefix_sep = ' = ')
simple_model_with_modellcsoport_text_vect = pd.get_dummies(simple_model_with_modellcsoport_text_vect, columns = categoricals_with_modellcsoport, prefix_sep = ' = ')
log_model_with_brand_text_vect = pd.get_dummies(log_model_with_brand_text_vect, columns = categoricals_with_marka, prefix_sep = ' = ')
log_model_with_modellcsoport_text_vect = pd.get_dummies(log_model_with_modellcsoport_text_vect, columns = categoricals_with_modellcsoport, prefix_sep = ' = ')
```

```python
%%time

# without text
level_brands = fit_sm_OLS(simple_model_with_brand)
log_brands = fit_sm_OLS(log_model_with_brand)
level_modells = fit_sm_OLS(simple_model_with_modellcsoport)
log_modells = fit_sm_OLS(log_model_with_modellcsoport)

# with text
level_brands_text = fit_sm_OLS(simple_model_with_brand_text)
log_brands_text = fit_sm_OLS(log_model_with_brand_text)
level_modells_text = fit_sm_OLS(simple_model_with_modellcsoport_text)
log_modells_text = fit_sm_OLS(log_model_with_modellcsoport_text)

# with text as vectors
level_brands_text_vect = fit_sm_OLS(simple_model_with_brand_text_vect)
log_brands_text_vect = fit_sm_OLS(log_model_with_brand_text_vect)
level_modells_text_vect = fit_sm_OLS(simple_model_with_modellcsoport_text_vect)
log_modells_text_vect = fit_sm_OLS(log_model_with_modellcsoport_text_vect)
```

Evaluating and comparing regression models

```python
ols_results_comparison = pd.concat(
          [OLS_results(level_brands, 'Level - brands'),
           OLS_results(level_modells, 'Level - modellcsoport'),
           OLS_results(log_brands, 'Log - brands'),           
           OLS_results(log_modells, 'Log - modellcsoport'),
           OLS_results(level_brands_text, 'Level - brands with text'),
           OLS_results(level_modells_text, 'Level - modellcsoport with text'),
           OLS_results(log_brands_text, 'Log - brands with text'),           
           OLS_results(log_modells_text, 'Log - modellcsoport with text'),
           OLS_results(level_brands_text_vect, 'Level - brands with text vect'),
           OLS_results(level_modells_text_vect, 'Level - modellcsoport with text vect'),
           OLS_results(log_brands_text_vect, 'Log - brands with text vect'),           
           OLS_results(log_modells_text_vect, 'Log - modellcsoport with text vect')])

ols_results_comparison.sort_values('r2')
```

The log-brands model seems best, as it is a lot simpler than any other, and still, performs at a similarly high level

## Build black box machine learning models
After simple OLS regressions let's check how much better we can get by using a blackbox ML model. Not going to use modellcsoport, as brands performed already at a high level, not going to use vectorized text, as one hot encoded version performed equally well and not going to use log models - tree based methods can take care of skewed targets. Will try
1. Random Forest
    - level 
        - brands
            - without text
            - text
2. GBM
    - same as for RF
    
    
3. CatBoost (keeping categoricals as categories, but not using text - Regressor does not yet support it)

#### Set data tables

```python
### one hot encoded data
# without text
simple_for_trees = simple_model_with_brand.copy()

# with text
with_text_for_trees = simple_model_with_brand_text.copy()

### original - categorical versions
original_for_catboost_no_text = regression_data[numerics + categoricals_with_marka + ['Vételár']].copy()
original_for_catboost_with_text = regression_data[numerics + categoricals_with_marka + felszereltseg + ['Vételár']].copy()
```

#### Train - test split

```python
%%time

simple_X = simple_for_trees.drop(['Vételár'], 1).copy()
simple_y = simple_for_trees['Vételár'].copy()
simple_X_train, simple_X_test, simple_y_train, simple_y_test = train_test_split(simple_X, simple_y, train_size = .8, random_state = 20202020)

simple_text_X = with_text_for_trees.drop(['Vételár'], 1).copy()
simple_text_y = with_text_for_trees['Vételár'].copy()
simple_text_X_train, simple_text_X_test, simple_text_y_train, simple_text_y_test = train_test_split(simple_text_X, simple_text_y, train_size = .8, random_state = 20202020)

catb_X = original_for_catboost_no_text.drop(['Vételár'], 1).copy()
catb_y = original_for_catboost_no_text['Vételár'].copy()
catb_X_train, catb_X_test, catb_y_train, catb_y_test = train_test_split(catb_X, catb_y, train_size = .8, random_state = 20202020)
```

### 1. Random Forest
#### A. without text

```python
%%time

RF_params = {'max_depth' : [15], 
             'min_samples_split' : [5],  
             'min_samples_leaf' : [10], 
             'max_features' : [1/2],
             'n_estimators' : [200]}

RF = RandomForestRegressor(random_state = 20202020, n_jobs = -1)

GRID_RF = GridSearchCV(RF, param_grid = RF_params, cv = 3, scoring = 'neg_root_mean_squared_error', n_jobs = -1)
GRID_RF.fit(simple_X_train, simple_y_train)
```

```python
rf_results = evaluate_model('RF simple', GRID_RF.best_estimator_, simple_X_test, simple_y_test)
```

#### B: with text

```python
%%time

RF_params = {'max_depth' : [20], 
             'min_samples_split' : [5],  
             'min_samples_leaf' : [10], 
             'max_features' : [1/2],
             'n_estimators' : [200]}

RF = RandomForestRegressor(random_state = 20202020, n_jobs = -1)

GRID_RF = GridSearchCV(RF, param_grid = RF_params, cv = 3, scoring = 'neg_root_mean_squared_error', n_jobs = -1)
GRID_RF.fit(simple_text_X_train, simple_text_y_train)
```

```python
rf_text_results = evaluate_model('RF with text', GRID_RF.best_estimator_, simple_text_X_test, simple_text_y_test)
```

### 2. GBM
#### A: without text

```python
%%time

GBM_params = {'min_samples_split': [5],
              'max_depth': [10],
              'max_features' : [0.5], 
              'subsample' : [0.75],
              'learning_rate' : [0.05],
              'n_estimators' : [300]}

GBM = GradientBoostingRegressor(random_state = 20202020)

GRID_GBM = GridSearchCV(GBM, param_grid = GBM_params, cv = 3, scoring = 'neg_root_mean_squared_error', n_jobs = -1)
GRID_GBM.fit(simple_X_train, simple_y_train) 
```

```python
gbm_results = evaluate_model('GBM simple', GRID_GBM.best_estimator_, simple_X_test, simple_y_test)
```

#### B: with text - didn't help random forest, so I won't bother trying it with GBM

### 3. CatBoost

```python
%%time

cat_params = {'learning_rate': [0.05],
              'l2_leaf_reg': [1],
              'subsample': [.8],
              'rsm' : [1/2],
              'max_depth': [15], # up to 16 
              'grow_policy': ['Lossguide'],
              'min_data_in_leaf' : [11], 
              'max_leaves' : [21],
              'iterations' : [1000]} 

cat = CatBoostRegressor(random_state = 20202020, one_hot_max_size = 30, eval_metric = 'RMSE', objective = 'RMSE',
                        cat_features = categoricals_with_marka, verbose = 0)

GRID_cat = GridSearchCV(cat, param_grid = cat_params, cv = 3, scoring = 'neg_root_mean_squared_error', n_jobs = -1)
GRID_cat.fit(catb_X_train, catb_y_train)
```

```python
catb_results = evaluate_model('CatBoost simple', GRID_cat.best_estimator_, catb_X_test, catb_y_test)
```

### BlackBox results

```python
bb_results = pd.concat([rf_results, rf_text_results, gbm_results, catb_results]).sort_values('r2')
bb_results
```

### Compare all

```python
pd.concat([ols_results_comparison, bb_results])[['Model', 'r2', 'features num']]
```

Tree models use simple level features and predict level (not log) price, and still, outperform log OLS models and greatly outperform level OLS regressions. Added bonus is the fact that for the tree based models the R2 is calculated on a separate test set - so completely new data. CatBoost is the best, probably because of the many categorical variables.
